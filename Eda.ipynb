KPI 1 internet cada 100 hogares trimestral


```python
acceso_x100H = pd.read_csv('Internet_Penetracion_x100_Hogares.csv')
acceso_x100H
```

```python
#dropear duplicados
acceso_x100H.drop_duplicates(inplace=True)
```
